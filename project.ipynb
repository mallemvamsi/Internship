```python
import pandas as pd
import numpy as np
from bokeh.plotting import output_file , show ,figure , ColumnDataSource
import matplotlib.pyplot as plt
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool
df=pd.read_csv("https://code.swecha.org/CivicTech/Indian-govt-funding-free-software/-/raw/patch-1/my_project.csv")
pd.set_option('display.max_rows',None)
df
```

```python
source=ColumnDataSource(df)
p=figure(plot_width=800,plot_height=800,title="FREE SOFTWARE PROJECTS FUNDED BY GOVERNMENT")
output_file("projects.html")
p.line(x="year",y="Total_outlay",source=source,line_width=4,color="green")
p.square(x="year",y="Total_outlay",source=source,line_width=4,color="yellow")
hover=HoverTool()
hover.tooltips=[('year','@year'),('TOTAL OUTLAY IN LAKHS','@Total_outlay'),('TITLE OF THE PROJECT','@Title_of_the_project'),('UNIVERSITY NAME','@Executing_Agency')]
p.add_tools(hover)
show(p)
```

```python
from bokeh.plotting import figure,output_file,show

output_file("govt.html")
p=figure(plot_width=1100,plot_height=500,title="CRIME IN HYDERABAD",y_axis_label="values",x_axis_label="year")
p.triangle(df["year"],df["Total_outlay"],size=14,line_color='cyan',fill_color='blue',fill_alpha=1,legend_label="TOTAL OUTLAY")

show(p)

 
```

```python
from math import pi
import pandas as pd
from bokeh.io import output_file, show
from bokeh.palettes import Category20c
from bokeh.plotting import figure
from bokeh.transform import cumsum
from bokeh.plotting import output_notebook
output_notebook()
output_file("GOVPIE2.html")
year=[]
outlay=[]
for i in range(0,89):
    year.append(str(df.year[i]))
    outlay.append(df.Total_outlay[i])

x=dict(zip(year,outlay))

data = pd.Series(x).reset_index(name='value').rename(columns={'index':'year'})
data['angle'] = data['value']/data['value'].sum() * 2*pi
data['color'] = Category20c[len(x)]

p = figure(plot_height=350, title="Pie Chart", toolbar_location=None,
           tools="hover", tooltips="@year: @value", x_range=(-0.5, 1.0))

p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend_field='year', source=data)

p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

show(p)
```
